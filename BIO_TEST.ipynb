```python
import json
import re

def extract_text_from_json(data):
    if isinstance(data, str):
        yield data
    elif isinstance(data, dict):
        for value in data.values():
            yield from extract_text_from_json(value)
    elif isinstance(data, list):
        for item in data:
            yield from extract_text_from_json(item)
    else:
        yield str(data)  # Convert non-string values to strings

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

# Load the JSON file
with open('vmbo1.json', 'r') as json_file:
    json_data = json.load(json_file)

# Extract all text values
all_text_values = list(extract_text_from_json(json_data))

# Remove HTML tags from each text value
cleaned_text_values = [remove_html_tags(text) for text in all_text_values]

# Print all cleaned text values
for text_value in cleaned_text_values:
    print(text_value)
```
